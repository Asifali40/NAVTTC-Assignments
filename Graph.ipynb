```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
#loadfile
df = pd.read_csv("/content/data.csv")
#plotting
plt.figure(figsize=(12,6))
plt.bar(df.index,df['Duration'],label= 'Duration' ,color = 'pink', alpha =0.7)
plt.bar(df.index, df['Pulse'],label='Pulse', color = 'green', alpha =0.7)
plt.bar(df.index, df['Maxpulse'],label='Maxpulse',color = 'red', alpha =0.7)
plt.bar(df.index, df['Calories'],label='Calories',color = 'orange', alpha =0.7)
plt.title('Fitness Data')
plt.xlabel('Index')
plt.ylabel('Values')
plt.legend()
plt.show()





```
